```python
# Removing the unwanted columns
dataset=dataset.drop(columns=["Person ID","Blood Pressure"])
```

```python
dataset.head(3)
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
dataset["Gender"] =pd.DataFrame(le.fit_transform(dataset["Gender"]))
dataset["Occupation"] =pd.DataFrame(le.fit_transform(dataset["Occupation"]))
dataset["BMI Category"] =pd.DataFrame(le.fit_transform(dataset["BMI Category"]))
dataset["Sleep Disorder"] =pd.DataFrame(le.fit_transform(dataset["Sleep Disorder"]))
```

```python
dataset.head(3)
```

```python
dataset["Sleep Disorder"].unique()
```

```python
# 2-> None
# Now split idata into input and output data
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Sleep Disorder"]
```

# Split data for training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=2000)
```

```python
lr.fit(x_train,y_train)
```

#Check score

```python
lr.score(x_train,y_train)*100,lr.score(x_test,y_test)*100
```
